```python
import pyautogui as pag
from pyperclip import copy, paste
from PIL import ImageGrab
from time import sleep
from more_itertools import windowed
from codecs import open
import json
```

```python
screenshot = lambda: ImageGrab.grab().load()
复制 = lambda: pag.click(538, 663)
下一题 = lambda: pag.click(897, 1358)
def 题目() -> str:
    pag.click(98, 97, 3, 0.2)
    pag.hotkey('ctrl', 'c')
    return paste()

def 长按(x, y, sec):
    pag.mouseDown(x, y)
    sleep(sec)
    pag.mouseUp()

def get_comments():
    im = screenshot()
    hots = []
    comments = []
    if im[74, 732] != (255, 255, 255):
        print("本题没有热门")
    else:
        hr_ys = []
        for y in range(739, 1308):
            prev = im[150, y]
            flag = False # 一致，True 不一致
            for x in range(151, 421):
                if im[x, y] != prev:
                    flag = True
                    break
            if not flag:
                if prev == (244, 244, 244):
                    # 没了
                    hr_ys.append(y)
                    break
                if prev == (238, 238, 238):
                    # hr
                    hr_ys.append(y)
        print("热门分割线:", hr_ys)
        for a, b in windowed(hr_ys, 2):
            长按(332, (a + b)/2, 1)
            sleep(.04)
            复制()
            hots.append(paste())
            sleep(.2)

    if im[660, 566] != (255, 255, 255):
        print("本题没有评论")
    else:
        hr_ys = []
        for y in range(581, 1308):
            prev = im[632, y]
            flag = False # 不一致
            for x in range(633, 947):
                if im[x, y] != prev:
                    flag = True
                    break
            if not flag:
                if prev == (244, 244, 244):
                    # 没了
                    hr_ys.append(y)
                    break
                if prev == (238, 238, 238):
                    # hr
                    hr_ys.append(y)
        for a, b in windowed(hr_ys, 2):
            长按(801, (a + b)/2, 1)
            sleep(.04)
            复制()
            comments.append(paste())
            sleep(.2)
    return hots, comments
```

```python
data = {}
for i in range(1, 972 + 1):
    sleep(1.2)
    print(f"==== 第 {i} 题 ====")
    hots, comments = get_comments()
    data[i] = {
        "title": 题目(),
        "hots": hots,
        "comments": comments
    }
    print(data[i])
    if i % 20 == 0:
        print("休息一下…")
        sleep(8)
    if i % 100 == 0:
        with open(f'data{i}.json', 'w', encoding='utf-8') as file:
            json.dump(data, file, indent=2, ensure_ascii=False)
        print("数据已储存。")
    下一题()

with open(f'data_fin.json', 'w', encoding='utf-8') as file:
    json.dump(data, file, indent=2, ensure_ascii=False)
print("数据已储存。")
```
